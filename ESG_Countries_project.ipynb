# 3.Data Cleaning

#### filling missing values using interpolation

```python
# Identify the year columns from 1970 to 2023
year_columns_1970_2023 = [str(year) for year in range(1970, 2024)]

# Apply interpolation to the year columns
df_interpolated = df_merged.copy()
df_interpolated[year_columns_1970_2023] = df_interpolated[year_columns_1970_2023].T.interpolate().T

display(df_interpolated.head())
```

```python
df_interpolated.info()
```

#### filling missing values using forward and backward filling

```python
year_columns_1970_2023 = [str(year) for year in range(1970, 2024)]
df_interpolated[year_columns_1970_2023] = df_interpolated[year_columns_1970_2023].fillna(method='ffill', axis=1).fillna(method='bfill', axis=1)
display(df_interpolated.head())
```

```python
display(df_interpolated.head())
display(df_interpolated.info())
```

```python
# Identify rows with missing values in 'ESG Pillar', 'Group', or 'Indicator name'
missing_categorical_mask = df_interpolated['ESG Pillar'].isna() | df_interpolated['Group'].isna() | df_interpolated['Indicator name'].isna()

# Drop the rows with missing categorical values
df_cleaned = df_interpolated[~missing_categorical_mask].copy()

display(df_cleaned.head())
display(df_cleaned.info())
```

# 6.Preparing data for data visualisation 2 (ESG scores)

#### Create a dictionary to store the scoring direction for each indicator based on whether a higher value is better or worse for ESG performance.


```python
indicator_scoring_direction = {
    'Level of water stress: freshwater withdrawal as a proportion of available freshwater resources': -1, # Higher water stress is worse
    'Total greenhouse gas emissions per capita excluding LULUCF (t CO2e/capita)': -1, # Higher emissions are worse
    'Renewable energy consumption (% of total final energy consumption)': 1, # Higher renewable energy consumption is better
    'Food production index (2014-2016 = 100)': 1, # Higher food production is better
    'Forest area (% of land area)': 1, # Higher forest area is better
    'GDP (annual % growth)': 1, # Higher GDP growth is generally better
    'School enrollment, primary and secondary (gross), gender parity index (GPI)': 1, # Closer to 1 is better for parity
    'Government Effectiveness: Estimate': 1, # Higher estimate is better
    'Voice and Accountability: Estimate': 1, # Higher estimate is better
    'Research and development expenditure (% of GDP)': 1, # Higher expenditure is better
    'Rule of Law: Estimate': 1, # Higher estimate is better
    'Access to electricity (% of population)': 1, # Higher access is better
    'Life expectancy at birth, total (years)': 1, # Higher life expectancy is better
    'Literacy rate, adult total (% of people ages 15 and above)': 1, # Higher literacy rate is better
    'Unemployment, total (% of total labor force) (modeled ILO estimate)': -1, # Higher unemployment is worse
    'Mortality rate, under-5 (per 1,000 live births)': -1, # Higher mortality rate is worse
    'Gini index': -1 # Higher Gini index indicates more inequality, which is worse
}
```

#### Normalize indicator values


Calculate the min and max values for each indicator and then normalize the values using min-max scaling, adjusting for the scoring direction.



```python
# Group by indicator and calculate min and max values
indicator_min_max = df_melted.groupby('Indicator name')['Value'].agg(['min', 'max']).reset_index()

# Merge min/max values back to the melted dataframe
df_melted = pd.merge(df_melted, indicator_min_max, on='Indicator name', how='left')

# Apply min-max normalization
df_melted['Normalized_Score'] = (df_melted['Value'] - df_melted['min']) / (df_melted['max'] - df_melted['min'])

# Adjust the normalized score based on the scoring direction
for indicator, direction in indicator_scoring_direction.items():
    if direction == -1:
        df_melted.loc[df_melted['Indicator name'] == indicator, 'Normalized_Score'] = 1 - df_melted.loc[df_melted['Indicator name'] == indicator, 'Normalized_Score']

display(df_melted.head())
display(df_melted.info())
```

```python
# Group by indicator and calculate min and max values
indicator_min_max = df_melted.groupby('Indicator name')['Value'].agg(['min', 'max']).reset_index()

# Merge min/max values back to the melted dataframe
df_melted = pd.merge(df_melted, indicator_min_max, on='Indicator name', how='left')

# Apply min-max normalization
df_melted['Normalized_Score'] = (df_melted['Value'] - df_melted['min_x']) / (df_melted['max_x'] - df_melted['min_x'])

# Adjust the normalized score based on the scoring direction
for indicator, direction in indicator_scoring_direction.items():
    if direction == -1:
        df_melted.loc[df_melted['Indicator name'] == indicator, 'Normalized_Score'] = 1 - df_melted.loc[df_melted['Indicator name'] == indicator, 'Normalized_Score']

display(df_melted.head())
display(df_melted.info())
```

#### Calculate weighted indicator scores


Define the indicator weights and calculate the weighted score for each indicator in the melted dataframe.



```python
# Define indicator weights (equal weights for now)
indicator_weights = {indicator: 1 for indicator in indicator_scoring_direction.keys()}

# Create a new column for Weighted_Score
df_melted['Weighted_Score'] = df_melted.apply(lambda row: row['Normalized_Score'] * indicator_weights[row['Indicator name']], axis=1)

display(df_melted.head())
```

#### Calculate esg pillar scores



Aggregate the weighted indicator scores by country, year, and ESG Pillar to get raw pillar scores.



```python
df_pillar_scores_raw = df_melted.groupby(['Economy_data', 'Year', 'ESG Pillar'])['Weighted_Score'].mean().reset_index()
display(df_pillar_scores_raw.head())
```

#### Normalize esg pillar scores


Group df_pillar_scores_raw by 'ESG Pillar' and calculate the minimum and maximum 'Weighted_Score' for each pillar, then merge these min/max values back into df_pillar_scores_raw to calculate the normalized pillar score using min-max normalization.



```python
# Group by 'ESG Pillar' and calculate the minimum and maximum 'Weighted_Score'
pillar_min_max = df_pillar_scores_raw.groupby('ESG Pillar')['Weighted_Score'].agg(['min', 'max']).reset_index()

# Merge these minimum and maximum values back into df_pillar_scores_raw
df_pillar_scores_normalized = pd.merge(df_pillar_scores_raw, pillar_min_max, on='ESG Pillar', how='left')

# Calculate the 'Normalized_Pillar_Score' using min-max normalization
df_pillar_scores_normalized['Normalized_Pillar_Score'] = (df_pillar_scores_normalized['Weighted_Score'] - df_pillar_scores_normalized['min']) / (df_pillar_scores_normalized['max'] - df_pillar_scores_normalized['min'])

# Drop the intermediate min and max columns
df_pillar_scores_normalized = df_pillar_scores_normalized.drop(columns=['min', 'max'])

display(df_pillar_scores_normalized.head())
display(df_pillar_scores_normalized.info())
```

Calculate combined esg score



Aggregate the normalized ESG Pillar scores by country and year to calculate the combined ESG score, then display the head of the resulting DataFrame.



```python
# Group by 'Economy_data' and 'Year' and calculate the mean of 'Normalized_Pillar_Score'
df_combined_esg = df_pillar_scores_normalized.groupby(['Economy_data', 'Year'])['Normalized_Pillar_Score'].mean().reset_index()

display(df_combined_esg.head())
```

# 7.Data Visualisation 2 (ESG Scores)

## Visualize individual esg pillar scores



Filter the data for individual pillar scores and plot each pillar's score over time for each country.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define a list of North African countries
north_african_countries = ['Algeria', 'Egypt, Arab Rep.', 'Libya', 'Morocco', 'Tunisia']

# Get unique ESG Pillars
unique_pillars = df_pillar_scores_normalized['ESG Pillar'].unique()
df_pillar_scores_normalized = df_pillar_scores_normalized[df_pillar_scores_normalized['Economy_data'].isin(north_african_countries)]
# Create a plot for each ESG Pillar
for pillar in unique_pillars:
    # Filter data for the current pillar
    pillar_df = df_pillar_scores_normalized[df_pillar_scores_normalized['ESG Pillar'] == pillar]

    # Create the line plot
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=pillar_df, x='Year', y='Normalized_Pillar_Score', hue='Economy_data')
    plt.title(f'{pillar} Pillar Score over Time in North Africa')
    plt.xlabel('Year')
    plt.ylabel('Normalized Score')
    plt.grid(True)
    plt.show()
```
